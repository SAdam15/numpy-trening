```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt

# Dane
cena_mieszkania_teraz = 120_000  # zł
inflacja_rocza = 0.05  # 5%
lata = 5
miesiecy = lata * 12

# Obliczenie przyszłej ceny mieszkania
cena_po_5_latach = cena_mieszkania_teraz * (1 + inflacja_rocza)**lata
print(f"Cena mieszkania za 5 lat: {cena_po_5_latach:.2f} zł")

# Parametry lokaty
stopa_roczna = 0.12
stopa_miesieczna = stopa_roczna / 12

# Obliczenie miesięcznej wpłaty, by uzbierać na mieszkanie
rata = -npf.pmt(stopa_miesieczna, miesiecy, 0, cena_po_5_latach)
print(f"Potrzebna miesięczna wpłata: {rata:.2f} zł")

# Wartości do wykresu
miesiace = np.arange(miesiecy + 1)
cena_miesiac = np.linspace(cena_mieszkania_teraz, cena_po_5_latach, miesiecy + 1)

# Stan lokaty w czasie
saldo = np.zeros(miesiecy + 1)
for i in range(1, miesiecy + 1):
    saldo[i] = saldo[i-1] * (1 + stopa_miesieczna) + rata

# Wykres
plt.figure(figsize=(10, 6))
plt.plot(miesiace, cena_miesiac, label="Cena mieszkania")
plt.plot(miesiace, saldo, label="Wartość lokaty")
plt.xlabel("Miesiące")
plt.ylabel("Kwota [zł]")
plt.title("Cena mieszkania vs. Wartość lokaty")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import numpy_financial as npf

# Parametry
cena = 120000                # Cena początkowa mieszkania
roczny_wzrost = 0.05         # Roczny wzrost cen mieszkań
lata = 5
okres_miesiecy = lata * 12

# Obliczenie przyszłej ceny mieszkania
przyszla_cena = cena * (1 + roczny_wzrost) ** lata

# Parametry lokaty
stopa_roczna = 0.12
stopa_miesieczna = stopa_roczna / 12

# Wyliczenie miesięcznej wpłaty potrzebnej do zgromadzenia tej kwoty
rata = npf.pmt(rate=stopa_miesieczna, nper=okres_miesiecy, pv=0, fv=przyszla_cena)

print(f"Cena mieszkania za 5 lat: {przyszla_cena:.2f} zł")
print(f"Potrzebna miesięczna wpłata: {-rata:.2f} zł")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Oś czasu: 60 miesięcy (5 lat)
miesiace = np.arange(1, okres_miesiecy + 1)

# Cena mieszkania rosnąca liniowo
cena_mieszkania = cena * (1 + roczny_wzrost) ** (miesiace / 12)

# Wartość lokaty w każdym miesiącu
wartosc_lokaty = npf.fv(rate=stopa_miesieczna, nper=miesiace, pmt=-rata, pv=0)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.plot(miesiace, cena_mieszkania, label='Cena mieszkania', color='red')
plt.plot(miesiace, wartosc_lokaty, label='Wartość lokaty', color='green')
plt.title('Porównanie ceny mieszkania i wartości lokaty (miesiąc po miesiącu)')
plt.xlabel('Miesiące')
plt.ylabel('Kwota [zł]')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
